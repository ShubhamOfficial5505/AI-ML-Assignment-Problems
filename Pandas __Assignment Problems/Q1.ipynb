```python
import pandas as pd
```

```python
iris_flower_df = pd.read_csv("IRIS.csv")
iris_flower_df
```

```python
# A
print(iris_flower_df.head(10))

print(iris_flower_df.shape)
print(iris_flower_df.dtypes)

print(iris_flower_df.describe().loc[["count", "std", "min", "max"]])
```

```python
# B
print(iris_flower_df[iris_flower_df["petal_length"] > 4.5])

print(iris_flower_df.query("species == 'Iris-virginica'")) # copy
```

```python
# C
iris_flower_df.groupby("species").agg(
    avg_sepal_length = ("sepal_length", "mean"),
    max_petal_width = ("petal_width", "max"),
    std_sepal_width = ("sepal_width", "std")
)
```

```python
# D
iris_flower_df_copy = iris_flower_df.copy()
iris_flower_df_copy = iris_flower_df_copy.assign(petal_ratio = (iris_flower_df_copy["petal_length"] / iris_flower_df_copy["petal_width"]))

iris_flower_df_copy.groupby("species")["petal_ratio"].mean()
```
