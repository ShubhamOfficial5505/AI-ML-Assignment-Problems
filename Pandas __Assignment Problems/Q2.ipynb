```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("Titanic-Dataset.csv")
titanic_df
```

```python
# A
titanic_df[["Name", "Sex", "Age", "Fare", "Survived"]]
```

```python
# B
print(titanic_df[titanic_df["Sex"] == "female"])

print(titanic_df.query("Fare > 30")) # copy
```

```python
# C
survived = titanic_df.groupby("Pclass")["Survived"].sum()
survived.name = None
survived.index.name = None

total_people = titanic_df.groupby("Pclass").size()
total_people.index.name = None

survival_rate = ((survived / total_people) * 100)

df_survival_rate = survival_rate.to_frame(name = "survival_rate")

df_pclass_info = titanic_df.groupby("Pclass").aggregate(
    average_fare = ("Fare", "mean"),
    average_age = ("Age", "mean")
)
df_pclass_info.index.name = None

df_pclass_info = pd.concat([df_pclass_info, df_survival_rate], axis = 1)
df_pclass_info.index.name = "Pclass"
df_pclass_info
```
